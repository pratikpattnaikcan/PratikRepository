To create the ingestion and retrieval workflow, we first need to install `chromadb` for our vector store and create a dummy HR policy document. We'll also ensure `langchain-google-genai` is installed for the embedding model.

Now, let's create a dummy HR policy document. In a real scenario, you would load your actual policy document from a file (e.g., PDF, TXT, DOCX).

```python
# Create a dummy HR policy document
hr_policy_content = """
# HR Policy Document: Employee Handbook

## 1. Introduction
Welcome to the company! This handbook provides an overview of our policies, procedures, and benefits. We aim to foster a positive and productive work environment.

## 2. Code of Conduct
Employees are expected to act professionally and ethically. Discrimination, harassment, and retaliation are strictly prohibited. Report any violations immediately.

## 3. Working Hours and Leave
Standard working hours are 9 AM to 5 PM, Monday to Friday. Employees are entitled to 20 days of paid annual leave per year, plus 10 public holidays. Sick leave requires a doctor's note for absences over 3 days.

## 4. Compensation and Benefits
Salaries are paid bi-weekly. Benefits include health insurance, dental, vision, and a 401(k) retirement plan with company matching. Full details are available through HR.

## 5. Performance Management
Regular performance reviews are conducted annually. Employees are encouraged to discuss career development and training opportunities with their managers.

## 6. Termination Policy
Employment can be terminated by either party with appropriate notice. Reasons for termination may include performance issues, misconduct, or redundancy. Details are outlined in individual employment contracts.

## 7. IT Policy
Company devices and networks are for business use. Proper data security practices must be followed. Unauthorized software installation is prohibited.

## 8. Expense Reimbursement
Employees can claim reimbursement for approved business expenses. Submit receipts within 30 days of the expense incurrence. Consult the finance department for guidelines.
"""

# Save the content to a file
with open("hr_policy.txt", "w") as f:
    f.write(hr_policy_content)

print("HR policy document 'hr_policy.txt' created successfully.")
```

Now, we'll set up the LangChain ingestion workflow:
1.  **Load the document:** We'll use `TextLoader` for our `.txt` file.
2.  **Split the document:** `RecursiveCharacterTextSplitter` will break the document into smaller, manageable chunks.
3.  **Create embeddings:** `GoogleGenerativeAIEmbeddings` will convert these text chunks into numerical vector representations.
4.  **Ingest into Chroma:** The chunks and their embeddings will be stored in a Chroma vector database.

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import Chroma
import os

# Ensure GOOGLE_API_KEY is set from Colab secrets (already handled by previous cells)

# 1. Load the document
loader = TextLoader("hr_policy.txt")
documents = loader.load()

# 2. Split the document into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
chunks = text_splitter.split_documents(documents)

# 3. Create embeddings
embeddings = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001")

# 4. Ingest into Chroma
# This will create a local ChromaDB instance
vectorstore = Chroma.from_documents(chunks, embeddings, persist_directory="./chroma_db")

print(f"Ingested {len(chunks)} chunks into ChromaDB.")
print("ChromaDB created at: ./chroma_db")
```

Since the embedding model `text-embedding-004` is also not found, let's list all available models to find a suitable embedding model. We can use the `google.generativeai` library directly to inspect available models.

```python
import google.generativeai as genai
import os

# Ensure GOOGLE_API_KEY is set (already handled by previous cells)

# Configure the genai library with your API key
genai.configure(api_key=os.environ.get("GOOGLE_API_KEY"))

print("Available embedding models:")
for m in genai.list_models():
    if "embedContent" in m.supported_generation_methods:
        print(m.name)
```

With the ingestion complete, let's set up the retrieval workflow. We'll create a retrieval chain that combines the Chroma vector store (as a retriever) with our Gemini Flash 2.5 LLM to answer questions based on the HR policy document.

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough, RunnableParallel

# Initialize the Gemini Flash 2.5 LLM for answering
llm_flash = ChatGoogleGenerativeAI(model="gemini-2.5-flash")

# Define a prompt template for the LLM to answer questions based on context
prompt = ChatPromptTemplate.from_template("""
Answer the following question based only on the provided context.
If the context does not contain the information, state that you cannot answer the question.

<context>
{context}
</context>

Question: {question}
""")

# Create a retriever from the vector store
retriever = vectorstore.as_retriever()

# Function to format documents for the prompt
def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

# Construct the retrieval chain using LCEL
rag_chain = (
    RunnableParallel({"context": retriever | format_docs, "question": RunnablePassthrough()})
    | prompt
    | llm_flash
    | StrOutputParser()
)

# Example query 1
question = "What are the standard working hours and annual leave days?"
response = rag_chain.invoke(question)

print(f"Question: {question}")
print(f"Answer: {response}")

print("\n---")

# Example query 2
question_2 = "How much is the salary for an entry-level software engineer?"
response_2 = rag_chain.invoke(question_2)
print(f"Question: {question_2}")
print(f"Answer: {response_2}")

print("\n---")

# New Example query 3
question_3 = "What is the company's policy on IT equipment and data security?"
response_3 = rag_chain.invoke(question_3)
print(f"Question: {question_3}")
print(f"Answer: {response_3}")
```
